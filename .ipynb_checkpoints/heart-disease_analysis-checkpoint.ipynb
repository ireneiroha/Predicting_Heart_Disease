```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("heart.csv")
```

```python
# Column descriptions (Data Dictionary) getting to know what each abbreviation means
data_description = {
    'age': 'Age of the person in years',
    'sex': 'Gender (1 = male, 0 = female)',
    'cp': 'Chest pain type (0: typical angina, 1: atypical, 2: non-anginal, 3: asymptomatic)',
    'trestbps': 'Resting blood pressure (in mm Hg)',
    'chol': 'Serum cholesterol (in mg/dl)',
    'fbs': 'Fasting blood sugar > 120 mg/dl (1 = true, 0 = false)',
    'restecg': 'Resting ECG results (0: normal, 1: ST-T abnormality, 2: left ventricular hypertrophy)',
    'thalach': 'Maximum heart rate achieved',
    'exang': 'Exercise induced angina (1 = yes, 0 = no)',
    'oldpeak': 'ST depression induced by exercise',
    'slope': 'Slope of peak exercise ST segment (0, 1, 2)',
    'ca': 'Number of major vessels colored by fluoroscopy (0–3)',
    'thal': 'Thalassemia (1 = normal, 2 = fixed defect, 3 = reversible defect)',
    'target': 'Heart disease (1 = has disease, 0 = no disease)'
}

# Display the dictionary in a readable way
for column, meaning in data_description.items():
    print(f"{column}: {meaning}")
```

```python
print("="*55)
print("STEP 1: DATA INSPECTION ")
print("="*55)
```

```python
#Describing the nature of the data in term of statistics(Count, mean,min,max and std)
print("\n Data description")
print(df.describe(include = "all"))
```

```python
print("=="*50)
print ("\n STEP 2 :DATA CLEANING")
print("=="*50)
```

```python
#checking on duplicates (found only one)
print(df.duplicated().sum()) 
```

```python
#droppin the one duplicate that was found
df = df.drop_duplicates()
```

```python

# checking how many duplicates are still their after removing them incase they remained
print("Duplicates after removal:", df.duplicated().sum())
# checking the new shape of our dataset after removing the duplicates(in terms of rows nad columns)
print("New shape of dataset:", df.shape)
```

```python

#Just checking if they will be any case of an outliers by describing each column 

print(df['age'].describe())
print(df['chol'].describe())
```

```python
print("=="*50)
print ("\n STEP 3 :DATA VISUALIZATION")
print("=="*50)
```

```python


plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='chol', hue='sex', palette='coolwarm')
plt.title("Age vs Cholesterol by Sex")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.legend(title="Sex (0=Female, 1=Male)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='chol', hue='target', palette='Set1')
plt.title("Cholesterol by Age and Heart Disease Status")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.legend(title="Heart Disease (1=Yes, 0=No)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(data=df, x='sex', y='chol', hue='age', palette='viridis')
plt.title("Cholesterol Distribution by Sex and Age")
plt.xlabel("Sex (0=Female, 1=Male)")
plt.ylabel("Cholesterol")
plt.legend(title="Age")
plt.show()
```

```python
sns.histplot(df['chol'], kde=True, bins=30, color='skyblue')
plt.title("Cholesterol Distribution")
plt.xlabel("Cholesterol")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='thalach', hue='target', palette='Set2')
plt.title("Age vs Max Heart Rate by Heart Disease")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate (thalach)")
plt.legend(title="Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='trestbps', hue='target', palette='coolwarm')
plt.title("Age vs Resting Blood Pressure by Heart Disease")
plt.xlabel("Age")
plt.ylabel("Resting BP (trestbps)")
plt.legend(title="Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='cp', y='chol', palette='pastel')
plt.title("Cholesterol Levels by Chest Pain Type")
plt.xlabel("Chest Pain Type")
plt.ylabel("Cholesterol")
plt.show()
```

```python
sns.boxplot(data=df, x='cp', y='chol')
plt.title("Cholesterol Levels by Chest Pain Type")
plt.xlabel("Chest Pain Type (0 = typical angina, 1 = atypical, 2 = non-anginal, 3 = asymptomatic)")
plt.ylabel("Serum Cholesterol (mg/dl)")
plt.show()
```
